# Qwen3 `<think>` injection demo

Does the chat template inject an **extra** `<think>` block when the assistant content already has one?

```python
from transformers import AutoTokenizer

tok = AutoTokenizer.from_pretrained("Qwen/Qwen3-0.6B", trust_remote_code=True)
```

## Case 1: Assistant content WITHOUT `<think>`

```python
msgs_no_think = [
    {"role": "user", "content": "What is 2+2?"},
    {"role": "assistant", "content": "The answer is 4."},
]
rendered = tok.apply_chat_template(
    msgs_no_think, tokenize=False, add_generation_prompt=False
)
print(rendered)
```

## Case 2: Assistant content WITH `<think>` already present

```python
msgs_with_think = [
    {"role": "user", "content": "What is 2+2?"},
    {
        "role": "assistant",
        "content": "<think>\n2+2 is basic addition.\n</think>\n\nThe answer is 4.",
    },
]
rendered = tok.apply_chat_template(
    msgs_with_think, tokenize=False, add_generation_prompt=False
)
print(rendered)
```

## Case 3: Multi-turn — earlier turns WITH `<think>`, last turn WITH `<think>`

```python
msgs_multi = [
    {"role": "user", "content": "What is 2+2?"},
    {"role": "assistant", "content": "<think>\nSimple math.\n</think>\n\n4"},
    {"role": "user", "content": "And 3+3?"},
    {"role": "assistant", "content": "<think>\nAlso simple.\n</think>\n\n6"},
]
rendered = tok.apply_chat_template(
    msgs_multi, tokenize=False, add_generation_prompt=False
)
print(rendered)
```
